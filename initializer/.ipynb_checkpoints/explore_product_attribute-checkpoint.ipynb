```python
import json
import pandas as pd
```

```python
fp = open("./source/item_master.json", "r")
content = json.loads(fp.read())
fp.close()
content
```

```python
checkdup = list()
unit_of_items = list()
for item in content:
    item['unit_of_items'] = item['unit_of_items'].replace("x", "*").replace('(', '').replace(")", "").split("*")[0]
    if item['unit_of_items'] not in checkdup:
        unit_of_items.append({"name": item['unit_of_items'], "slug": None})
        checkdup.append(item['unit_of_items'])
unit_of_items
checkdup
```

```python
checkdup = list()
for item in content:
    attr = item['unit_of_items'].replace("x", "*")
    if "*" in attr:
        checkdup.append(attr)
        
print(json.dumps(checkdup))
```

```python
fp = open("./master_product_attribute.json", "w")
fp.write(json.dumps(unit_of_items, indent=4, ensure_ascii=False))
fp.close()
```

```python
def get_attr(name, data):
    for d in data:
        if name.replace("x", "*").replace('(', '').replace(")", "").split("*")[0] == d['name']:
            return d

choice = list()
for item in content:
    attr = get_attr(item['UNIT_OF_ITEMS'], unit_of_items)
    temp = dict()
    temp['name'] = f"{attr['name']}*{item['UNIT_FACT']}"
    temp['color'] = ''
    temp['attribute'] = attr['id']
    temp['slug'] = None
    if temp not in choice:
        choice.append(temp)
choice
```

```python
fp = open("./master_attribute_choice.json", "w")
fp.write(json.dumps(choice, indent=4, ensure_ascii=False))
fp.close()
```
