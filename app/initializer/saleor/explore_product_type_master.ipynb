```python
fp = open("./item_master.json", "r")
items = json.loads(fp.read())
fp.close()

items
```

```python
total_len = len(items)
total_len
```

```python
barcode_len = len(set([item['BARCODE'] for item in items]))
barcode_len
```

```python
stock_bar_len = len(set([item['STOCK_CODE']+'-'+item['BARCODE'] for item in items]))
stock_bar_len
```

```python
itemcode_len = len(set([item['ITEM_CODE'] for item in items]))
itemcode_len
```

```python
attributes = ['name', 'has_variants', 'is_shipping_required']

product_types = list()
for item in items:
    temp = dict()
    temp['name'] = item['SUBCATERY_NAME']
    temp['has_variants'] = False
    temp['is_shipping_required'] = True
    if temp not in product_types:
        product_types.append(temp)

product_types_frame = pd.DataFrame(product_types)
product_types_frame
```

```python
fp = open("./master_product_type.json", "w")
fp.write(json.dumps(product_types, indent=4, ensure_ascii=False))
fp.close()
```
